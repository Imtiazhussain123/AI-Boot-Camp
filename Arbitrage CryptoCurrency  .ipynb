```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
import numpy



a = requests.get('https://api.cryptonator.com/api/full/btc-usd').json()
b = a['ticker']['markets']
df = pd.DataFrame(b)

x = df['price'].astype(float)
y = df['volume']
z = df['market']


# print(y)
for i, txt in enumerate(z):
    plt.annotate(txt,(x[i],y[i]))

max_price = max(x)
min_price = min(x)
max_mrkt =max(z)
min_mrkt=min(z)
print(max_mrkt,'have the maximum price of',max_price)
print(min_mrkt,'have the minimum price of',min_price)
#Percent Difference  =  abs(One Value - Another Value)      x 100%
 #                       abs((One Value + Another Value)/2))


diff_price = max_price - min_price
print("The difference in "+max_mrkt+' and '+min_mrkt+' is '+ str(diff_price))
percentages =((diff_price)/((min_price)/2))*100 
print('Percentage'+'{:04.2f}%'.format(percentages))

plt.scatter(x,y)
plt.xlabel('Price')
plt.ylabel('Volume')
plt.title('Arbitage Graph')
plt.grid()
plt.show()


```
